#### What products are most often sold together?

```python
df.head()
```

```python
ef = df[df["Order ID"].duplicated(keep=False)]
ef.head(30)
```

```python
ef["Grouped"] = ef.groupby("Order ID")["Product"].transform(lambda x:",".join(x))
ef = ef[["Order ID", "Grouped"]].drop_duplicates()
```

```python
ef.head(100)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in ef["Grouped"]:
    row_list = row.split(",")
    count.update(Counter(combinations(row_list, 2)))

#print(count)

count.most_common(10)
```

```python
product_group = df.groupby("Product")
quantity_ordered = product_group.sum()["Quantity Ordered"]

products = [product for product, all_data in product_group]
```

```python
prices = df.groupby("Product").mean()["Price Each"]

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered)
ax2.plot(products, prices, "g")

ax1.set_xlabel("Products name")
ax1.set_xlabel("Quantity Ordered", color = "g")
ax2.set_xlabel("Prices", color = "g")
ax1.set_xticklabels(products, rotation="vertical", size = 8)

plt.show()
```
