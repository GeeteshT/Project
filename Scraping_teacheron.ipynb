# **Part 2**

```
# Now we are extracting from the page No-2, Here we have multiple name of the tutor so we are getting the list of all the tutor and the link of each tutor

url= urlopen("https://www.teacheron.com/tutors-in-india?p=2").read()

soup=BeautifulSoup(url,"html.parser")

soup
```

```
# h stands for header

h=soup.find("div",class_="row margin-bottom-30 highlightDivOnHover")
h
```

```
# Getting the single link from the webpage

url_link=soup.find("div",class_="row margin-bottom-30 highlightDivOnHover")

url_link.find("a",target="_blank")["href"]
```

```
# Geeting the list of all url 

url_list=[]
for col in soup.find_all("div",class_="row margin-bottom-30 highlightDivOnHover"):
  url=col.find("a",target="_blank")["href"]
  url_list.append(url)

print(url_list)
```

```
# Now we are scaping all the data
ew_name=[]
new_description=[]
subject_list=[]
experience_list=[]
education_list=[]
fees_list=[]
for col in soup.find_all("div",class_="row margin-bottom-30 highlightDivOnHover"):
  url=col.find("a",target="_blank")["href"]
  new_url=urlopen(url).read()
  h=BeautifulSoup(new_url,"html.parser")
  g=h.find("div",class_="wrapper")
  name=g.find("h1",class_="margin-bottom-40").span.text
  new_name.append(name)
  description=h.find("div",class_="profile-body1").span.text
  new_description.append(description)
  for col in h.find_all("h4",class_="no-margin"):
    subject_list.append(col.text)
  experience=h.find("div",class_="panel-body margin-bottom-10").get_text()  #https://www.teacheron.com/tutor-profile/28sG #https://www.teacheron.com/tutor-profile/287V
  experience_list.append(experience.strip())
  education=h.find("div",class_="panel-body padding-bottom-0").div.text
  education_list.append(education.split("\n")[1])
  money=h.find_all("ul",class_="list-unstyled vertical-list col-sm-12")[-1].p.text
  fees_list.append(money.strip())
```

```
# Creating the DataFrame
tutor_data=pd.DataFrame(list(zip(new_name,new_description,subject_list,experience_list,education_list,fees_list)),columns=["name","description","subject","Experience","Education","Fees"]) 
```
